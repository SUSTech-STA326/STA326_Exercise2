```python
# Imports
import requests  # send request
from bs4 import BeautifulSoup  # parse web pages
import pandas as pd  # csv
from time import sleep  # wait
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import time
```

```python
# Define a request header (to prevent anti-scraping)
headers = {
    'authority': 'curlconverter.com',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'accept-language': 'zh-CN,zh;q=0.9,en;q=0.8,en-GB;q=0.7,en-US;q=0.6',
    'cache-control': 'max-age=0',
    'if-modified-since': 'Fri, 15 Jul 2022 21:44:42 GMT',
    'if-none-match': 'W/"62d1dfca-3a58"',
    'referer': 'https://curlconverter.com/',
    'sec-ch-ua': '" Not A;Brand";v="99", "Chromium";v="102", "Microsoft Edge";v="102"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Linux"',
    'sec-fetch-dest': 'document',
    'sec-fetch-mode': 'navigate',
    'sec-fetch-site': 'cross-site',
    'sec-fetch-user': '?1',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.63 Safari/537.36 Edg/102.0.1245.30',
}
 # 把请求伪装成正常浏览器的请求
```

```python
url='https://arxiv.org/list/cs/pastweek?skip=0&show=100'
page=requests.get(url=url,headers=headers)
soup=BeautifulSoup(page.text, 'html.parser')
```

```python
title=[]
subject=[]
author=[]
abstract=[]
href=[]
n=len(soup.find_all(class_='list-title mathjax'))
for i in range(n):
    title.append(soup.find_all(class_='list-title mathjax')[i].text.strip()[7:])
    subject.append(soup.find_all(class_='list-subjects')[i].text.strip()[10:])
    author.append(soup.find_all(class_='list-authors')[i].text.strip()[9:].split(', \n'))
    href.append('https://arxiv.org'+soup.find_all(class_='list-identifier')[i].a['href'])

def find_abstract(next_page_url):
    # next_page_url ='https://arxiv.org/abs/2403.00762'
    response_next = requests.get(next_page_url,headers=headers)
    subsoup = BeautifulSoup(response_next.text, 'html.parser')
    next_page_title = subsoup.find_all(class_='abstract mathjax')[0].text.strip()
    return next_page_title[9:]

for i in range(n):
    time.sleep(2)  # 暂停2秒
    print('current page: ',i)
    abstract.append(find_abstract(next_page_url=href[i]))
```

```python
df=pd.DataFrame({'title':title,
              'subject':subject,
              'author':author,
              'href':href,
              'abstract': abstract})
df
```
