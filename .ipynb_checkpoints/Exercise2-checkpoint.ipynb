```python
import requests  
from bs4 import BeautifulSoup  
import pandas as pd  
from time import sleep 
import numpy as np
```

```python
headers = {
    'authority': 'curlconverter.com',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'accept-language': 'zh-CN,zh;q=0.9,en;q=0.8,en-GB;q=0.7,en-US;q=0.6',
    'cache-control': 'max-age=0',
    'if-modified-since': 'Fri, 15 Jul 2022 21:44:42 GMT',
    'if-none-match': 'W/"62d1dfca-3a58"',
    'referer': 'https://curlconverter.com/',
    'sec-ch-ua': '" Not A;Brand";v="99", "Chromium";v="102", "Microsoft Edge";v="102"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Linux"',
    'sec-fetch-dest': 'document',
    'sec-fetch-mode': 'navigate',
    'sec-fetch-site': 'cross-site',
    'sec-fetch-user': '?1',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.63 Safari/537.36 Edg/102.0.1245.30',
}
```

```python
url="https://arxiv.org/list/cs/pastweek?skip=0&show=100"
page=requests.get(url,headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
subjects = [] 
subject_elements = soup.select('div.meta div.list-subjects')
for subject_element in subject_elements:
    subject = subject_element.text.strip().replace('Subjects:', '')
    subjects.append(subject)
```

```python
titles = []  
title_elements = soup.select('div.meta div.list-title.mathjax')
for title_element in title_elements:
    title = title_element.text.strip().replace('Title:', '')
    titles.append(title)
```

```python
authors=[]
author_elements = soup.select('div.meta div.list-authors')
for author_element in author_elements:
    author = author_element.text.strip()
    author=author.replace('Authors:', '').replace('\n', '').strip()
    authors.append(author)
```

```python
abstracts=[]
link_elements = soup.select('span.list-identifier a[title="Abstract"]')
for link_element in link_elements:
    url2="https://arxiv.org/"+link_element['href']
    page2=requests.get(url2,headers=headers)
    soup2 = BeautifulSoup(page2.content, 'html.parser')
    abstract=soup2.select_one('.abstract.mathjax')
    abstract=abstract.text.strip().replace("Abstract:",'')
    abstracts.append(abstract)
```

```python
paper={'Title':titles,'Subjects':subjects,'Authors':authors,'Abstract':abstracts}
paper_df=pd.DataFrame(paper)
paper_df
```

```python
paper_df.to_csv("D://first_100_papers.csv",index=False)
```
