```python
import requests  # send request
from bs4 import BeautifulSoup  # parse web pages
import pandas as pd  # csv
from time import sleep  # wait
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import re
```

```python
proxies = {
    "http": "http://127.0.0.1:7890",
    "https": "http://127.0.0.1:7890"
}

url = "https://arxiv.org/list/cs/recent?show=100"
url = "https://arxiv.org/search/advanced?advanced=1&terms-0-operator=AND&terms-0-term=&terms-0-field=title&classification-computer_science=y&classification-physics_archives=all&classification-include_cross_list=exclude&date-filter_by=all_dates&date-year=&date-from_date=&date-to_date=&date-date_type=submitted_date&abstracts=show&size=100&order=-announced_date_first"
response = requests.get(url = url, proxies = proxies)
```

```python
# print(response.content)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
# print(soup)
```

```python
list_title = soup.find_all("p", class_ = "title is-5 mathjax")
list_authors = soup.find_all("p", class_ = "authors")
list_subjects = soup.find_all("div", class_ = "tags is-inline-block")
list_abstracts = soup.find_all("p", class_ = "abstract mathjax")

titles = []
authors = []
subjects = []
abstracts = []

for title, author, subject, abstract in zip(list_title, list_authors, list_subjects, list_abstracts):
    abstract = re.search(r'▽ More\n\s*(.*?)\n\s*△ Less', abstract.text, re.DOTALL).group(1).strip()
    author = re.sub(r'\s+', ' ',author.text.replace("Authors:", "").replace("\n", "").replace("\t", "").strip())
    title = title.text.strip()  
    subject = ", ".join(re.findall(r'data-tooltip="(.*?)"', str(subject)))
    abstracts.append(abstract)
    titles.append(title)
    authors.append(author)
    subjects.append(subject)
#     print(
#         abstract,
#         title,
#         author,
#         subject
#     )
```

```python
df = pd.DataFrame({
    "Title": titles,
    "Authors": authors,
    "Subject": subjects,
    "Abstract": abstracts
})

df.to_csv("./out.csv", index = False)
```
