```python
soups = []
url = f"https://arxiv.org/list/cs/pastweek?skip=0&show=100"
response = requests.get(url,headers=headers)
soup =BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
abs_url = soup.find_all('a', title='Abstract')
```

```python
import csv
import pandas as pd

data_list = []

for i in range(len(abs_url)):
# for i in range(5):
    paper_index = abs_url[i]['href']
    url = f"https://arxiv.org{paper_index}"
    print(url)
    response_paper = requests.get(url, headers=headers)
    soup_paper = BeautifulSoup(response_paper.content, 'html.parser')

    title = soup_paper.find_all('h1', class_='title mathjax')[0].text[6:]
    subject = soup_paper.find_all('div', class_='subheader')[0].text[1:-1]
    authors = soup_paper.find_all('div', class_='authors')[0].text[8:]
    abstract = soup_paper.find_all('blockquote', class_='abstract mathjax')[0].text[10:].replace('"', '')

    paper_data = {'title': title, 'subject': subject, 'authors': authors, 'abstract': abstract}
    data_list.append(paper_data)

df = pd.DataFrame(data_list)

df.to_csv('arxiv_100.csv', index=False, encoding='utf-8')

print("CSV file 'arxiv_100.csv' has been created.")
```
